```python
# open the output file: it is 2 columns of numbers
# the first column is the number of elements
# the second column is the time in seconds

# read the file
f = open('output.txt', 'r')
lines = f.readlines()
f.close()

# create two empty lists
x = []
y = []

# fill the lists with values from the file nut the last one
for line in lines[:-1]:
    p = line.split()
    x.append(float(p[0]))
    y.append(float(p[1]))


```

```python

import matplotlib.pyplot as plt 



# plot log only on x axis
plt.plot(x, y)
plt.xscale('log')
plt.xlabel('Number of vertices')
plt.ylabel('Memory usage (MB)')
plt.title('Memory usage by number of vertices')
plt.show()
```
